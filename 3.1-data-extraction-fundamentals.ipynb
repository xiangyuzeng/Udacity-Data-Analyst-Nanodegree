```python
def parse_file(datafile):
    data = []
    with open(datafile, "r") as f:
        firstline = True
        for line in f:
            if firstline:
                firstline = False
                continue
            else:
                parsed_line = [el.strip() for el in line.split(',')]
                line_dict = {}
                line_dict['Title'] = parsed_line[0]
                line_dict['Released'] = parsed_line[1]
                line_dict['Label'] = parsed_line[2]
                line_dict['UK Chart Position'] = parsed_line[3]
                line_dict['US Chart Position'] = parsed_line[4]
                line_dict['BPI Certification'] = parsed_line[5]
                line_dict['RIAA Certification'] = parsed_line[6]
                if len(data) < 10:
                    data.append(line_dict)
    return data
```

```python
def parse_file_ans(datafile):
    data = []
    with open(datafile, 'r') as f:
        header = f.readline().split(',')
        counter = 0
        for line in f:
            if counter == 10:
                break
            fields = line.split(',')
            dict_entry = {}
            for i, value in enumerate(fields):
                dict_entry[header[i].strip()] = value.strip()
            data.append(dict_entry)
            counter +=1
    return data
```

```python
def xlrd_example(datafile):
    workbook = xlrd.open_workbook(datafile) # Open a workbook
    sheet = workbook.sheet_by_index(0) # Specify sheet to work with. xlrd uses zero-based indexing.
    
    # List-based comprehension: 
    # Looping through rows and columns 
    # and reading that data into a Python list
    data = [[sheet.cell_value(r, col) 
                for col in range(sheet.ncols)] 
                    for r in range(sheet.nrows)]
    
    print("\nList Comprehension")
    print("data[3][2]:",) 
    # Print Row 3 Col 2
    print(data[3][2])

    print("\nCells in a nested loop:" )   
    for row in range(sheet.nrows):
        for col in range(sheet.ncols):
            if row == 50:
                # Print all values in that row, one col at a time
                print(sheet.cell_value(row, col),)


    ### other useful methods:
    print("\nROWS, COLUMNS, and CELLS:")
    print("Number of rows in the sheet:", )
    print(sheet.nrows)
    print("Type of data in cell (row 3, col 2):",) 
    print(sheet.cell_type(3, 2))
    print("Value in cell (row 3, col 2):",) 
    print(sheet.cell_value(3, 2))
    print("Get a slice of values in column 3, from rows 1-3:")
    print(sheet.col_values(3, start_rowx=1, end_rowx=4)) # Upper bound non-inclusive

    print("\nDATES:")
    #Prev dates in Excel represented as floating points
    print("Type of data in cell (row 1, col 0):",) 
    print(sheet.cell_type(1, 0))
    exceltime = sheet.cell_value(1, 0)
    print("Time in Excel format:",) 
    print(exceltime)
    print("Convert time to a Python datetime tuple, from the Excel float:",)
    print(xlrd.xldate_as_tuple(exceltime, 0)) # Helps us work with date in Python

xlrd_example('2013_ERCOT_Hourly_Load_Data.xls')
```

```python
# -*- coding: utf-8 -*-
'''
Find the time and value of max load for each of the regions
COAST, EAST, FAR_WEST, NORTH, NORTH_C, SOUTHERN, SOUTH_C, WEST
and write the result out in a csv file, using pipe character | as the delimiter.

An example output can be seen in the "example.csv" file.
'''

import xlrd
import os
import csv
from zipfile import ZipFile

datafile = "2013_ERCOT_Hourly_Load_Data.xls"
outfile = "2013_Max_Loads.csv"


def open_zip(datafile):
    with ZipFile('{0}.zip'.format(datafile), 'r') as myzip:
        myzip.extractall()


def parse_file(datafile):
    # YOUR CODE HERE
    # Remember that you can use xlrd.xldate_as_tuple(sometime, 0) to convert
    # Excel date to Python tuple of (year, month, day, hour, minute, second)
    
    workbook = xlrd.open_workbook(datafile)
    sheet = workbook.sheet_by_index(0)
    data = []
    
    header = ["Station","Year","Month","Day","Hour","Max Load"]
    for i in range(0,sheet.ncols):
        # Omit first row because it is a header row
        if i == 0:
            continue
        else:
            station = sheet.cell_value(0,i)
            station_values = sheet.col_values(i, start_rowx=1, end_rowx=sheet.nrows)
            max_load_value = max(station_values)
            max_load_pos = station_values.index(max_load_value) + 1
            max_load_time_xls = sheet.cell_value(max_load_pos,i)
            max_load_time = xlrd.xldate_as_tuple(max_load_time_xls, 0)
            row_data = [station, max_load_time[0], max_load_time[1], max_load_time[2], max_load_time[3], max_load_value]
            data.append(row_data)
    return data

def save_file(data, filename):
    # YOUR CODE HERE
    csv_writer = csv.writer(open(outfile, 'wb'), delimiter='|')
    for i in range(0,len(data)):
        csv_writer.writerow(data[i])
    
def test():
    open_zip(datafile)
    data = parse_file(datafile)
    save_file(data, outfile)

    number_of_rows = 0
    stations = []

    ans = {'FAR_WEST': {'Max Load': '2281.2722140000024',
                        'Year': '2013',
                        'Month': '6',
                        'Day': '26',
                        'Hour': '17'}}
    correct_stations = ['COAST', 'EAST', 'FAR_WEST', 'NORTH',
                        'NORTH_C', 'SOUTHERN', 'SOUTH_C', 'WEST']
    fields = ['Year', 'Month', 'Day', 'Hour', 'Max Load']

    with open(outfile) as of:
        csvfile = csv.DictReader(of, delimiter="|")
        for line in csvfile:
            station = line['Station']
            if station == 'FAR_WEST':
                for field in fields:
                    # Check if 'Max Load' is within .1 of answer
                    if field == 'Max Load':
                        max_answer = round(float(ans[station][field]), 1)
                        max_line = round(float(line[field]), 1)
                        assert max_answer == max_line

                    # Otherwise check for equality
                    else:
                        assert ans[station][field] == line[field]

            number_of_rows += 1
            stations.append(station)

        # Output should be 8 lines not including header
        assert number_of_rows == 8

        # Check Station Names
        assert set(stations) == set(correct_stations)

        
"""if __name__ == "__main__":
    test()"""
```
